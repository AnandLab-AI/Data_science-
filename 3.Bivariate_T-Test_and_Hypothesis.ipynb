```python
from scipy.stats import ttest_ind
male=dataset[dataset['hsc_s']=='M']['salary']
Female=dataset[dataset['hsc_s']=='F']['salary']
ttest_ind(male,Female)
```

```python
## T value 1.90 

#0 → very strong evidence (big difference)
#1 → no evidence (no difference)

## P-Value 0.05

# p < 0.05 → ✅ Significant
# p ≥ 0.05 → ❌ Not significant

#| Test        | When to use                                |
#| ----------- | ------------------------------------------ |
#| `ttest_ind` | 2 different groups (Male vs Female)        |
#| `ttest_rel` | Same people (before vs after, SSLC vs HSC) |
```

```python
## Paried T Test
##Same group (male) different condition 

from scipy.stats import ttest_rel
male=dataset[dataset['gender']=='M']['ssc_p']
male1=dataset[dataset['gender']=='M']['etest_p']
ttest_ind(male,male1)
```

```python
## p Value 
## p = 2.92e-08 that means 0.0000000292  -- > so p < 0.05 p-value is extremely small
## t-value = -4.32 Big magnitude → strong difference , Negative sign → second group mean > first group mean

## -- > There is a statistically significant difference between the two groups, and the second group has higher values than the first group.
```

```python
##Same group (commerece ) different condition  ( mba_p , etest_p)

from scipy.stats import ttest_rel
com1=dataset[dataset['hsc_s']=='Commerce']['mba_p']
com2=dataset[dataset['hsc_s']=='Commerce']['etest_p']
ttest_rel(com1,com2)
```

```python
##There is a highly statistically significant difference between MBA marks and Entrance test scores for Commerce students.
## E-test scores > MBA scores
```

# HYPOTHESIS


```python
###  Test whether MBA mark of commerce and entrance test make of commerce student make is same ?
### Null Hypothesis (H₀) ( or) Alternative Hypothesis (H₁)

from scipy.stats import ttest_rel
com1=dataset[dataset['hsc_s']=='Commerce']['mba_p']
com2=dataset[dataset['hsc_s']=='Commerce']['etest_p']
ttest_rel(com1,com2)
```

```python
## p-value = 2.46e-12 =  2.46e-12 < 0.05 
## There is a statistically significant difference between MBA and entrance test scores.
""" Since p-value < 0.05, we reject the null hypothesis. There is a statistically significant difference between MBA 
and entrance test scores for Commerce students, and entrance test scores are higher than MBA scores."""
```
