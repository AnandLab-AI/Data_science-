```python
quan=[]
qual=[]

for colname in dataset.columns:
    if dataset[colname].dtypes == "O":
        qual.append(colname)
    else:
        quan.append(colname)
```

```python
quan ,qual
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(dataset[quan])
t_data=imp_mean.transform(dataset[quan])
```

```python
t_data
```

```python
t_data=pd.DataFrame(t_data,columns=quan)
```
