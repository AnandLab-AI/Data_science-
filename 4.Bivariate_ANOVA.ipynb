```python
from scipy.stats import f_oneway
commerce = dataset[dataset['hsc_s'] == 'Commerce']['mba_p']
science = dataset[dataset['hsc_s'] == 'Science']['mba_p']
arts = dataset[dataset['hsc_s'] == 'Arts']['mba_p']

f_oneway(commerce, science, arts)
```

```python
#| Item      | Meaning                   |
#| --------- | ------------------------- |
#| p = 0.766 | No significant difference |
#| F = 0.266 | Very low variation        |
#| Result    | All groups similar        |

#| F value               | Meaning              |
#| --------------------- | -------------------- |
#| Small F (≈ 1 or less) | groups are similar   |
#| Large F               | groups are different |

"""F is very small , So groups are almost same , No strong difference """

""" p = 0.766  -- > 0.766 > 0.05 ===== #So :  We fail to reject H₀"""

## We do not have evidence to say groups are different, so we keep the Null Hypothesis.
```

## Two Way ANOVA

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example: Dependent variable = mba_p
# Factors = hsc_s (High School Stream), gender

# Build the model
model = ols('mba_p ~ C(hsc_s) + C(gender) + C(hsc_s):C(gender)', data=dataset).fit()

# Perform Two-Way ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)



# mba_p → Dependent variable (MBA percentage).
# C(hsc_s) → Factor 1 (High School stream, categorical).
# C(gender) → Factor 2 (Gender, categorical).
# C(hsc_s):C(gender) → Interaction effect (does stream effect depend on gender?).
# anova_lm(model, typ=2) → Gives ANOVA table with:
# sum_sq → Sum of squares (variance explained).
# df → Degrees of freedom.
# F → F‑statistic.
# PR(>F) → p‑value (significance).
```

```python
## Gender → significant factor influencing MBA percentage.
## High School Stream → not significant.
## Interaction (HSC Stream × Gender) → not significant.

## So, for your dataset, gender is the only factor that matters statistically for MBA percentage. Stream doesn’t add explanatory power.
```
