```python
descriptive=pd.DataFrame(index=("Mean","Median","Mode"),columns=quan)
for colname in quan:
    descriptive[colname]["Mean"]=dataset[colname].mean()
```

```python
## Error when we run descriptive[colname]["Mean"]=dataset[colname].mean()


##❌ Problem with chained indexing
##descriptive[colname] returns a temporary object (copy or view)
##Then ["Mean"] tries to assign a value on that temporary object
##Pandas cannot guarantee whether the original DataFrame is updated
##This can lead to:
##Incorrect results
##Or warnings like SettingWithCopyWarning

## .loc is a label-based indexing method in Pandas.
## DataFrame.loc[row_label, column_label]
```

```python
descriptive=pd.DataFrame(index=("Mean","Median","Mode"),columns=quan)
for colname in quan:
    descriptive.loc["Mean",colname]=dataset[colname].mean()
```

```python
descriptive
```

```python
descriptive=pd.DataFrame(index=("Mean","Median","Mode"),columns=quan)
for colname in quan:
    descriptive.loc["Mean",colname]=dataset[colname].mean()
    descriptive.loc["Median",colname]=dataset[colname].median()
```

```python
descriptive
```

```python
descriptive=pd.DataFrame(index=("Mean","Median","Mode"),columns=quan)
for colname in quan:
    descriptive.loc["Mean",colname]=dataset[colname].mean()
    descriptive.loc["Median",colname]=dataset[colname].median()
    descriptive.loc["Mode",colname]=dataset[colname].mode()[0]
```
